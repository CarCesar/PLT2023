```python
texto = '''Sergipe – Fábio Mitidieri (PSD) – R$ 39 mil (salário vai ser reajustado para R$ 46,3 mil em 2025)
Acre – Gladson Cameli (Progressistas) – R$ 35,8 mil
Mato Grosso do Sul – Eduardo Riedel (PSDB) – R$ 35,4 mil
Rio Grande do Sul – Eduardo Leite (PSDB) – R$ 35,4 mil
Rondônia – Marcos Rocha (União) – R$ 35,4 mil
Pará – Helder Barbalho (MDB) – R$ 35,3 mil
Bahia – Jerônimo Rodrigues (PT) – R$ 35 mil
São Paulo – Tarcísio de Freitas (Republicanos) – R$ 34,5 mil
Roraima – Antonio Denarium (PP) – R$ 34,2 mil
Amazonas – Wilson Lima (União) – R$ 34 mil
Paraná – Ratinho Júnior (PSD) – R$ 33,7 mil
Amapá – Clécio Luis (Solidariedade) – R$ 33 mil
Paraíba – João Azevêdo (PSB) – R$ 31,1 mil
Piauí – Rafael Fonteles (PT) – R$ 30,2 mil (salário vai ser reajustado para R$ 31,2 mil)
Mato Grosso – Mauro Mendes (União) – R$ 29,4 mil
Espírito Santo – Renato Casagrande (PSB) – R$ 29,4 mil
Alagoas – Paulo Dantas (MDB) – R$ 27,7 mil
Goiás – Ronaldo Caiado (União) – R$ 27,5 mil
Santa Catarina – Jorginho Mello (PL) – R$ 25,3 mil
Tocantins – Wanderlei Barbosa (Republicanos) – R$ 24,1 mil (salário vai ser reajustado para R$ 28 mil)
Distrito Federal – Ibaneis Rocha (MDB) – R$ 24 mil
Pernambuco – Raquel Lyra (PSDB) – R$ 22 mil
Rio Grande do Norte – Fátima Bezerra (PT) – R$ 21,9 mil
Rio de Janeiro – Claudio Castro (PL) – R$ 21,8 mil
Ceará – Elmano de Freitas (PT) – R$ 19,4 mil
Maranhão – Carlos Brandão (PSB) – R$ 15,9 mil
Minas Gerais – Romeu Zema (Novo) – R$ 10,5 mil'''
```

```python
texto = texto.split('\n')
```

```python
lista = []
for a in texto:
  lista.append(a.split(' – '))

lista
```

```python
for a in range(len(lista)):
  temp1 = lista[a][1].replace(')','').split(' (')
  lista[a][1]=temp1[0]
  lista[a].append(temp1[1])
  temp2 = lista[a][2].replace(',','.').split(' mil')
  lista[a][2]=float(temp2[0].replace('R$',''))
```

```python
lista
```

```python
import pandas as pd
```

```python
Code = pd.read_html('https://www.oobj.com.br/bc/article/quais-os-c%C3%B3digos-de-cada-uf-no-brasil-465.html')[0].rename(columns={'Código UF':'Code', 'Unidade da Federação':'UF', 'UF':'XX'})
Code.loc[22,'UF'] = 'Rio Grande do Sul'
Code
```

```python
dados = pd.DataFrame(lista,columns=['UF','Gov','Sal','Par'])
dados = dados.merge(Code)
dados
```

```python
dados.columns
```

```python
import altair as alt
```

```python
from altair.vegalite.v4 import StrokeWidth
terra = alt.topo_feature('https://raw.githubusercontent.com/CarCesar/PLT2023/main/BR.json','BRUF')
fundo = alt.Chart(terra).mark_geoshape(stroke="black", strokeWidth=0.5,fill="white")

click1 = alt.selection_multi(fields=['UF'])

mapa1 = alt.Chart(terra,height=650,width=680
         ).mark_geoshape(
                stroke="black", 
                strokeWidth=0.15,
         ).transform_lookup(
                lookup = 'properties.codarea',
                from_ = alt.LookupData(
                                        data=dados, 
                                        key='Code', 
                                        fields=['UF', 'Gov', 'Sal', 'Par', 'XX'])
         ).encode(
             color=alt.Color('Sal:Q',legend=None),
             opacity=alt.condition(click1,alt.value(1),alt.value(0.1)),
             tooltip = [alt.Tooltip('UF:N'),alt.Tooltip('Gov:N'), alt.Tooltip('Sal:N'), alt.Tooltip('Par:N'), 'XX:N']
         ).add_selection(click1)

barra = alt.Chart(dados,height=650).mark_bar(clip=True).encode(
    x=alt.X('Sal',title='Salário(mil Reais)',scale=alt.Scale(domain=[0,40],zero=False)),
    y=alt.Y('XX',title='UF',sort='-x'),
    color=alt.Color('Sal',legend=None),
    opacity=alt.condition(click1,alt.value(1),alt.value(0.1)),
    tooltip = [alt.Tooltip('UF:N'),alt.Tooltip('Gov:N'), alt.Tooltip('Sal:N'), alt.Tooltip('Par:N'), 'XX:N']
    ).add_selection(click1)

(mapa1|barra).properties(title='Salário dos Governadores (Março de 2023)').configure_title(anchor='middle',fontSize=28).configure_view(strokeWidth=0)#.project(type='equirectangular',).properties(width=680, height=650,title='Minas Gerais pelos Partidos dos prefeitos (2020)')
```
